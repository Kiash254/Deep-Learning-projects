```python
#importing the Libraries That are supposed to be used in Image processing and Classification

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
# loading an image through matplotlib.image module

img = mpimg.imread('/content/dog.jpg')
```

```python
#check the type of  the image using type
type(img)
```

```python
#print the shape of the image
img.shape
```

```python
#print the image to see the arrays
print(img)
```

```python
# displaying the image from numpy array
plot=plt.imshow(img)
plt.title('Image showing a Puppy in a RGB')
plt.show()
```

```python
#now i want to resize the image using the pillow library
from PIL import Image
img=Image.open('/content/dog.jpg')
img_resized = img.resize((200, 200))
```

```python
img_resized.save('dog_image_resized.jpg')
```

```python
# displaying the image from numpy array
img_res = mpimg.imread('/content/dog_image_resized.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

```python
#print the shape to check the shape of the resized image
print(img_res.shape)
```

[link text](https://)**Converting RGB images to Grayscale image using OpenCV**

```python
# importing OpenCV library
import cv2
```

```python
img = cv2.imread('/content/dog.jpg')
```

```python
#check the type  of  image
type(img)
```

```python
#check the shape of the image
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
grayscale_image.shape
```

cv2.imshow() will display the image. But this will not be allowed in Google Colab.

from google.colab.patches import cv2_imshow

```python
from google.colab.patches import cv2_imshow
# displaying the image
cv2_imshow(grayscale_image)
```

```python
# saving the grayscale image
cv2.imwrite('dog_grayscale_image.jpg', grayscale_image)
```
