```python
from google.colab import files
from transformers import pipeline
from PIL import Image
from io import BytesIO

# Load the image-to-text pipeline
pipe = pipeline("image-to-text", model="Salesforce/blip-image-captioning-large")

# Upload an image file
uploaded = files.upload()

# Get the image bytes from the uploaded file
image_bytes = uploaded[next(iter(uploaded))]

# Convert the image bytes to a PIL image object
img = Image.open(BytesIO(image_bytes))

# Extract the text from the image using the pipeline
result = pipe(img)

# Print the result dictionary
print(result)
```

```python
from google.colab import files
from PIL import Image
from transformers import pipeline
from io import BytesIO

# Load the object detection pipeline
pipe = pipeline("object-detection", model="facebook/detr-resnet-50")

# Upload an image file
uploaded = files.upload()

# Get the image bytes from the uploaded file
image_bytes = uploaded[next(iter(uploaded))]

# Convert the image bytes to a PIL image object
img = Image.open(BytesIO(image_bytes))

# Perform object detection on the image using the pipeline
result = pipe(img)

# Print the detected objects and their coordinates
for obj in result:
    print(obj["label"], obj["score"], obj["box"])
```
